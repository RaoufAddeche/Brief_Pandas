```python
import pandas as pd


# Charger les fichiers
categories_df = pd.read_csv('castorama_categories.csv')
products_df = pd.read_csv('castorama_products.csv')

# Afficher les premières lignes
print("Catégories :")
print(categories_df.head())
print("Produits :")
print(products_df.head())
```

```python
#doublons dans les titres
products_df['title'].describe()
```

```python
# les doublons sont du au faits que plusieurs produits ont le meme titre mais pas les memes marques
nb_title = pd.DataFrame(products_df.groupby('title')['title'].count())


nb_title[nb_title['title'] > 1]
```

```python
#list des categories

products_df['category'].unique()
```

```python
# Dimensions des DataFrames
print("Dimensions des catégories :", categories_df.shape)
print("Dimensions des produits :", products_df.shape)

# Vérification des valeurs manquantes
print("Valeurs manquantes dans catégories :")
print(categories_df.isna().sum())

print("Valeurs manquantes dans produits :")
print(products_df.isna().sum())
```

```python
# Trier les produits par titre
sorted_products = products_df.sort_values(by='title')
print("Produits triés par titre :")
print(sorted_products.head(10))
```

```python
# Filtrer les produits appartenant à la catég quincaillerie
quincaillerie_products = products_df[products_df['category'] == 'quincaillerie']
print("Produits dans la catégorie 'quincaillerie' :")
print(quincaillerie_products.head(10))
```

```python
# Sauvegarder les produits filtrés
quincaillerie_products.to_csv('quincaillerie_products.csv', index=False)
print("sauvegarder")
```

```python
# Nettoie la la colonne prix, remplace les , par des . et remplace tout n'est pas numérique par du vide que ce soit au début et à la fin

products_df['price'] = products_df['price'].str.replace(',', '.')
products_df['price'] = products_df['price'].str.replace(' ', '')
products_df['price'] = products_df['price'].str.replace('(None,)', '0' )
products_df['price'] = products_df['price'].str.replace('(None.)', '0' )
```

```python
# Convertir les prix en format numériq dans le fichier produits (ils sont actuellement en texte).

products_df['price'] = products_df['price'].astype(float)
```

```python
# Suppression des lignes egales à 0
products_df = products_df[products_df['price'] != 0]

products_df
```

```python
print(products_df.isna().sum())  # Identifier les valeurs manquantes
products_df = products_df.dropna(subset=['price'])  #supprimer les produits sans prix
```

```python
products_df['category'] = products_df['category'].str.lower()  #passe tout en minuscule
products_df = products_df.drop_duplicates()  # Supprimer les doublons
```

```python
print(products_df['price'].describe())
```

```python
#Compter le nombre de produits par catégorie
#Calculer le prix moyen des produits pour chaque catégorie

print(products_df['category'].value_counts())  # Nombre de produits par catégorie
avg_price_per_category = products_df.groupby('category')['price'].mean()
print(avg_price_per_category)
```

```python
#Relier les fichiers categories_df et products_df
merged_df = pd.merge(products_df, categories_df, on='category', how='left')
print(merged_df.head())
```

```python
#colonne indiquant si une catégorie est une "page list".
products_df = pd.merge(products_df, categories_df[['category', 'is_page_list']], on='category', how='left')
print(products_df.head())

```
