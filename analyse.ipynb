```python
# Nettoie la la colonne prix, remplace les , par des . et remplace tout n'est pas numérique par du vide que ce soit au début et à la fin

df_products['price'] = df_products['price'].str.replace(',', '.')
df_products['price'] = df_products['price'].str.replace(' ', '')
df_products['price'] = df_products['price'].str.replace('(None,)', '0' )
df_products['price'] = df_products['price'].str.replace('(None.)', '0' )
```

```python
#modifie le type de la colonne price
df_products['price'] = df_products['price'].astype(float)
```

```python
# Suppression des lignes egales à 0
df_products = df_products[df_products['price'] != 0]

df_products
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Exemple : Calculer les différences de prix
price_differences = df_quincaillerie['price'].diff().dropna()

# Calculer les paramètres de la loi normale (moyenne et écart-type)
mu, sigma = np.mean(price_differences), np.std(price_differences)

# Générer les points pour la courbe
x = np.linspace(min(price_differences), max(price_differences), 1000)
y = norm.pdf(x, mu, sigma)  # Fonction de densité de probabilité

# Tracer l'histogramme des différences
plt.figure(figsize=(10, 6))
plt.hist(price_differences, bins=30, density=True, alpha=0.6, color='skyblue', label='Histogramme des différences')

# Ajouter la courbe de loi normale
plt.plot(x, y, color='red', linewidth=2, label=f'Loi normale ($\mu={mu:.2f}$, $\sigma={sigma:.2f}$)')

# Ajouter des titres et des légendes
plt.title('Courbe de loi normale des différences de prix', fontsize=16)
plt.xlabel('Différences de prix', fontsize=14)
plt.ylabel('Densité de probabilité', fontsize=14)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Afficher le graphique
plt.tight_layout()
plt.show()
```
